```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./books.csv")
df
```

```python
ds = ['title', 'authors', 'original_publication_year', 'average_rating', 'ratings_count', 'small_image_url','books_count']
df = df[ds]
```

```python
harrypoter_books = df[ df['title'].str.contains('Harry Potter')]
harrypoter_books
```

```python
max_index = harrypoter_books['books_count'].idxmax() 
max_value = df.loc[max_index, 'books_count'] 
title_value = df.loc[max_index, 'title']   
print("Maximum value:", max_value)
print("Title :", title_value)
```

```python
average = (harrypoter_books.average_rating * harrypoter_books.ratings_count).sum() / harrypoter_books.ratings_count.sum()
average
```
